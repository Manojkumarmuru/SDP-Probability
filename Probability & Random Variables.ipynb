### Coin Flip Simulator Using Random Number Generation

```python
import random as rand
import matplotlib.pyplot as plt
```

### The Coin Flip Experiment
Flipping of coin can be a random experiment. The side of the coin that shows up on the top after flipping the coin can be considered as the random variable $X$. <br>
Then the flipping of coin becomes an experiment that results in either a head or a tail. In the below implementtain, the random variable $X=1$ is onsidered to be Heads and $X=0$ is considered to be tail. <br>
The frequency of occurance of heads and tails as a resulto fthe random experiment is shown as a bar graph.

```python
outcomes = list()    # A list to append the outcome
trials = 100         # Setting the number of trials

for trial in range(0, trials):
    
    flip = rand.randint(0,1)           # Pick randomly 0 or 1
                                       # Assumption-1    
    if flip:                           # 1 represents heads
        outcomes.append("Heads")
    else:                              # 0 represents tails
        outcomes.append("Tails")

# Plot the frequency of occurance of heads and tails as bar chart
heads_count = outcomes.count("Heads")
tails_count = outcomes.count("Tails")
plt.bar(["Frequence of Heads", "Frequency of Tails"], [heads_count, tails_count])
plt.text(0,heads_count+1,heads_count)
plt.text(1,tails_count+1,tails_count)
plt.title("Frequency of Heads and Tails from Random Coin Tossing Experiment")
plt.show()
```

### Plotting Probability Mass Functions for Discrete Random Variables

```python
import scipy.stats as st
```

#### Bernoulli Discrete Random Variable

A Bernoulli random variable$(X)$ takes two values namely $X=1$ and $X=0$ with a probability of occurance as $p$ and $1-p$ respectively.

\begin{equation}
p(X)=
\begin{cases}
  1-p  & \text{for } X=0 \\
  p  & \text{for } X=1
\end{cases}
\end{equation}

<br>
In the example code given below the probability of occurance of $X=1$ is taken to be $p=0.7$ and so the probability of occurance of $X=0$ is $0.3$

```python
p = 0.7
random_var = st.bernoulli(p)
plt.bar([-1, 0, 1, 2], [random_var.pmf(-1), random_var.pmf(0), random_var.pmf(1), random_var.pmf(2)])
plt.text(0, random_var.pmf(0) + 0.01, round(random_var.pmf(0), 2))
plt.text(1, random_var.pmf(1) + 0.01, round(random_var.pmf(1), 2))
plt.title("Bernoulli Discrete Random Variable")
```

### Binomial Discrete Random Variable

A discrete binomial random variable is modelled using two parameters namely $n$ and $p$ which denote the number of times an independent experiment is carried out and the probability of success of an event respectively. <br>The number of times $(k)$ success is achieved out of performing a random experiment is taken to be a random variable $X$. Therefore this random variable can be given by the following distribution. <br>

$$
p(X=k)=
\left(\begin{array}{cc} 
n\\
k
\end{array}\right)
\left(\begin{array}{cc} 
p^{k}
\end{array}\right)
(1-p)^{n-k}
$$ 
<br>
The probability distribution of discrete Binomial random variable is plotted as a bar chart as shown below.

```python
p = 0.5
n = 10
random_var = st.binom(n, p)

probabilities = list()
for random_idx in range(-1, n+2):
    probabilities.append(random_var.pmf(random_idx))

plt.bar(range(-1, n+2), probabilities)
plt.title("Binomial Discrete Random Variable")
```

### Plotting Probability Density Functions for Continuous Random Variables

```python
import numpy as np
```

#### Alpha Continuous Random Variable

The alpha continuous random is denoted by the following probability density function, <br><br>
$f(x)=\frac{1}{2\pi}\int^{\infty}_{-\infty}\varphi(t)e^{-ixt}dt$ <br><br>
Where $\varphi(t; \alpha; \beta, c, \mu  = exp(it\mu - |ct|^{\alpha}(1-i\beta sgn(t)\phi))$ <br><br>
Where, $\phi$ is given as, <br>
\begin{equation}
\phi=
\begin{cases}
  tan(\frac{\pi \alpha}{2})  & \alpha \neq1 \\
  -\frac{2}{\pi}log|t|  & \alpha =1
\end{cases}
\end{equation}

```python
a = 0.5
random_var = st.alpha(a)

x = np.arange(-2.5, 10, 0.05)
plt.plot(x, random_var.pdf(x))
plt.title("Alpha Continuous Random Variable")
```

#### Cauchy Continuous Random Variable

The probability density function of a continuous cauchy distribution is given as, <br><br>
$f(x; x_{0}, \gamma)=\frac{1}{\pi\gamma [1+\frac{x-x_{0}}{\gamma}]^{2}}$ <br><br>
where, $x_{0}$ and $\gamma$ are the location and scale parameter respectively.

```python
x_0 = 0
gamma = 2
random_var = st.cauchy(x_0, gamma)

x = np.arange(-30, 30, 0.05)
plt.plot(x, random_var.pdf(x))
plt.title("Cauchy Continuous Random Variable")
```

### Plotting Probability Density Functions for Multivariate Random Variables

#### Multivariate Normal Random Variable

The probability mass function of a multivariate random variable is given as,<br><br>
$\phi(x)=(\frac{1}{2\pi})^{p/2}|\Sigma|^{-1/2}exp{-\frac{1}{2}(x-\mu)^{'}\Sigma^{-1}(x-\mu)}$ <br><br>
Where $\mu$ is the mean and $\Sigma$ is the singular matrix.

```python
mean = 20
cov = 50
random_var = st.multivariate_normal(mean, cov)

x = np.arange(-mean, 3*mean, 0.05)
plt.plot(x, random_var.pdf(x))

plt.title("Multivariate Normal Random Variable")
```

#### Multinomial Random Variable

The probability mass function of a multinomial random variable is given as, <br><br>

\begin{equation}
f(x_{1}, ..., x_{k}; p_{1},...p_{k})=
\begin{cases}
  \frac{n!}{x_{1}!...x_{k}!}p_{1}^{x_{1}}*...*p_{k}^{x_{k}}, when  \Sigma^{k}_{i=1}x_{i}=n\\
0, otherwise \\
\end{cases}
\end{equation}

```python
n = 5
p=[10]
random_var = st.multinomial(n, p)

pmfs = list()
xs = list()
for x in range(n-4, n+5, 1):
    pmfs.append(random_var.pmf([x]))
    xs.append(x)

plt.plot(xs, pmfs)
plt.title("Multinomial Random Variable")
```
